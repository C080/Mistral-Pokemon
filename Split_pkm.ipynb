```python
import json
import requests
response = requests.get('https://replay.pokemonshowdown.com/gen8randombattle-{}.json'.format('2143065686'))
if response.status_code == 200:
    battle = response.json()
    with open('pkmlog.json', 'w') as f:
        json.dump(battle, f)

players = {
        'p1': {},
        'p2': {}
           }
for row in battle['log'].split('\n'):
    if '|switch|p1a:' in row:
        pkm = row[12:].strip().split('|')[1]
        pkm = pkm.split(', ')
        if pkm[0] not in players['p1'].keys():
            players['p1'][pkm[0]]={'liv': pkm[1], 'mosse': []}

       #mosse cosa accadrà? le riempiamo con le mosse che verranno utilizzate
       #  
    elif '|switch|p2a:' in row:
        pkm = row[12:].strip().split('|')[1]
        pkm = pkm.split(', ')
        if pkm[0] not in players['p2'].keys():
            
            players['p2'][pkm[0]]={'liv': pkm[1], 'mosse': []}
```

```python

for row in battle['log'].split('\n'):
    if '|move|' in row: 
        list = row[6:].split('|')
   
        if 'p1a' in list[0]:
            p = 'p1'
        else: p = 'p2'
        pkm = list[0][5:]
        for pkms in players[p]:
            if pkm in pkms and list[1] not in players[p][pkms]['mosse']:
                players[p][pkms]['mosse'].append(list[1])
```
