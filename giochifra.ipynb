```python
## leggiamo il file dei log della battaglia
file = open("pkm_test.txt", "r", encoding='UTF-8')
file_letto = file.readlines()
len(file_letto)
file.close()
file_letto
```

```python
turni = []  # Lista dei turni (0,19)
maxturn = max([int(riga.strip('|turn|').strip()) for riga in file_letto if '|turn|' in riga])  # Lista dei turni presenti nel file di log
    
for i in range(maxturn): #fallo in generale non solo per log che hanno 21 turni
    turno = []  # Lista per memorizzare le righe del turno corrente
    for riga in file_letto:
        turno.append(riga)
        if f'|turn|{i+1}' in riga:  # Controlla se la riga contiene l'indicatore del turno corrente
            turno.pop(-1)  # Rimuove l'ultima riga aggiunta, che contiene l'indicatore del turno
            break
    turni.append(turno)  # Aggiunge il turno corrente alla lista dei turni
```

```python
# Scrivi ciascun turno in un file separato
for i, turno in enumerate(turni):
    nome_file = f"turno_{i}.txt"  # Nome del file basato sull'indice del turno
    with open(nome_file, 'w', encoding='UTF-8') as f:
        f.writelines(turno)
```
